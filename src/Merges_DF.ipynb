```python
# Transformamos la fecha a formato fecha
```

```python
calendar['date'] = pd.to_datetime(calendar['date'],format = '%Y-%m-%d')
```

# Convertimos las columnas del DF de SALES en filas, para poder hacer merge por la columna 'd' con el DF de CALENDAR

```python
# Lista de columnas a mantener fijas
fixed_columns = ['id','item', 'category', 'department', 'store', 'store_code', 'region']

# Aplicar la función melt para convertir las columnas d_1 a d_1913 en filas
sales_melt = pd.melt(sales, id_vars=fixed_columns, value_vars=sales.columns[6:], var_name='d', value_name='ventas',ignore_index=True)
```

```python
sales_melt
```

# Merge df_calendar y sales_melt

```python
df_calendar_sales = pd.merge(sales_melt,calendar, on =  'd',how = 'inner')
```

```python
df_calendar_sales.drop(['d','weekday_int','weekday'],axis = 1, inplace=True)
df_calendar_sales
```

# Agrupación de ventas semanalmente

```python
df_calendar_sales_by_id = (
    df_calendar_sales
    .set_index('date')
    .groupby(['id','item','category','department','store','store_code','region'])
    .resample('W')
    .agg(
        {
          "ventas" : np.sum,
          "event" : 'count',
        }
    ).reset_index()
)

df_calendar_sales_by_id
```

```python
df_ventas_semanales = df_calendar_sales_by_id.copy()
```

```python
df_ventas_semanales['event'].sum()
```

```python
df_ventas_semanales['year'] = df_ventas_semanales['date'].dt.year
df_ventas_semanales['week_of_year'] = df_ventas_semanales['date'].dt.strftime("%W")
df_ventas_semanales
```

```python
print('year min:', df_ventas_semanales['year'].min(), 'year max:', df_ventas_semanales['year'].max())
print('week_of_year min:', df_ventas_semanales['week_of_year'].min(), 'week_of_year max:', df_ventas_semanales['week_of_year'].max())
```

# Merge DF df_ventas_semanales y price

```python
print(df_ventas_semanales.shape)

df_ventas_semanales = pd.merge(df_ventas_semanales,prices,on = ['item','category','store_code','year','week_of_year'],how = 'left')

print(df_ventas_semanales.shape)
```

# Imputación de valores nulos

```python
df_ventas_semanales["sell_price_imputed"] = df_ventas_semanales.groupby(['id','item', 'category', 'department', 'store', 'store_code', 'region'])["sell_price"].transform(
    lambda series: series.backfill().ffill())
```

```python
df_ventas_semanales.isnull().sum()
```

```python
df_ventas_semanales[df_ventas_semanales['ventas'] == 0]
```

```python
df_ventas_semanales[(df_ventas_semanales['ventas'] == 0) & (df_ventas_semanales['sell_price'].isnull())]
```

```python
df_ventas_semanales.sort_values(by=['id','date'],inplace=True)
```

```python
cantidad_fechas = df_ventas_semanales.groupby('id')['date'].nunique()
```

```python
max(cantidad_fechas != 274)
```

```python
df_final_fechas = df_ventas_semanales.copy()
```

```python
df_final_fechas['revenue'] = df_final_fechas['ventas']*df_final_fechas['sell_price_imputed']
```

```python
df_final_fechas[df_final_fechas['region'] == 'Philadelphia']
```

```python
df_final_fechas.drop('sell_price',axis=1,inplace=True)
```

```python
df_final_fechas.rename(columns = {"sell_price_imputed":"price"},inplace=True)
```
